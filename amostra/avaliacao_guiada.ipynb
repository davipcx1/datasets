## Qual o percentual de cada tipo de imóvel na nossa base de dados?

```python
dados['Tipo']
```

```python
dados['Tipo'].value_counts()
```

```python
dados['Tipo'].value_counts(normalize=True)
```

```python
type(dados['Tipo'].value_counts(normalize=True))
```

```python
df_percentual_tipo = dados['Tipo'].value_counts(normalize=True).to_frame().sort_values('Tipo')
```

```python
df_percentual_tipo = dados['Tipo'].value_counts(normalize=True).to_frame().sort_values('Tipo')
```

```python
df_percentual_tipo.plot(
    kind='bar',                # tipo de gráfico (era "Rind", deve ser "kind")
    figsize=(14, 10),          # tamanho do gráfico
    color='green',             # cor das barras (era 'gree')
    edgecolor='black',         # borda das barras
    xlabel='Tipo',             # rótulo do eixo X
    ylabel='Percentual',       # rótulo do eixo Y
    title='Distribuição Percentual por Tipo',  # título do gráfico
    legend=False               # oculta a legenda, se quiser
)
```

```python
df_percentual_tipo.plot(Rind='bar', figsize=(14,10), color ='gree', edgecolor='black',
            xlabel = 'Tipo'
```
